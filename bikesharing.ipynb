## 1. Descriptive Analysis

### A. Calculate the **mean** of the hourly number of order(s)

```python
avg_n_order = df['total_order'].mean()
print(f'MEAN of the hourly number of order: {avg_n_order}')
```

### B. Calculate the **median** of the hourly number of order(s)

```python
mid_n_order = df['total_order'].median()
print(f'MEDIAN of hourly number of order: {mid_n_order}')
```

### C. Calculate the **mode** of the hourly number of order(s)

```python
mode_n_order = df['total_order'].mode()
print(f'MODE of hourly number of order: {mode_n_order}')
```

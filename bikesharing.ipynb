## 1. Descriptive Analysis

### A. Plot the total hourly number of order(s) histogram

```python
x = df['total_order']
fig = plt.hist(x, bins=20)
plt.title("Hourly N of Order(s) Histogram")
plt.xlabel('Hourly N of Order(s)')
plt.xticks(range(0,1100,100))
plt.show()
```

### B. Calculate the **mean** of the hourly number of order(s)

```python
avg_n_order = df['total_order'].mean()
avg_n_order = round(avg_n_order, 2)
print(f'MEAN of the hourly number of order: {avg_n_order}')

x = df['total_order']
fig = plt.hist(x, bins=20)
plt.title("Hourly N of Order(s) Histogram")
plt.xlabel('Hourly N of Order(s)')
plt.xticks(range(0,1100,100))
plt.axvline(x=np.nanmean(x), color='red', linestyle='--', label='Mean') # Mean line
plt.legend()
plt.show()
```

### C. Calculate the **median** of the hourly number of order(s)

```python
mid_n_order = df['total_order'].median()
print(f'MEDIAN of hourly number of order: {mid_n_order}')

x = df['total_order']
fig = plt.hist(x, bins=20)
plt.title("Hourly N of Order(s) Histogram")
plt.xlabel('Hourly N of Order(s)')
plt.xticks(range(0,1100,100))
plt.axvline(x=np.nanmedian(x), color='black', linestyle='--', label='Median') # Median line
plt.legend()
plt.show()
```

### D. Calculate the **mode** of the hourly number of order(s)

```python
mode_n_order = df['total_order'].mode()[0] # Get the mode
print(f'MODE of hourly number of order: {mode_n_order}')

x = df['total_order']
fig = plt.hist(x, bins=20)
plt.title("Hourly N of Order(s) Histogram")
plt.xlabel('Hourly N of Order(s)')
plt.xticks(range(0,1100,100))
plt.axvline(x=mode_n_order, color='black', linestyle='-', label='Mode') # Mode line
plt.legend()
plt.show()
```

### E. **Recap** of the Hourly n of Order Data Centrality

```python
print('Mean:', avg_n_order)
print('Median:', mid_n_order)
print('Mode:', mode_n_order)

x = df['total_order']
fig = plt.hist(x, bins=20)
plt.title("Hourly N of Order(s) Histogram")
plt.xlabel('Hourly N of Order(s)')
plt.xticks(range(0,1100,100))
plt.axvline(x=mode_n_order, color='black', linestyle='-', label='Mode') # Mode line
plt.axvline(x=np.nanmedian(x), color='black', linestyle='--', label='Median') # Median line
plt.axvline(x=np.nanmean(x), color='red', linestyle='--', label='Mean') # Mean line
plt.legend()
plt.show()
```

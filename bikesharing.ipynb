## 1. Descriptive Analysis

### A. Hourly Total Order Distribution and Centrality

```python
avg_hourly_n_order = round(df['total_order'].mean(), 2) # Get the mean
mid_hourly_n_order = df['total_order'].median() # Get the median
mode_hourly_n_order = df['total_order'].mode()[0] # Get the mode

x = df['total_order']
fig = plt.hist(x, bins=50)
plt.title("Hourly # of Order(s) Distribution")
plt.xlabel('Hourly # of Order(s)')
plt.xticks(range(0,1100,100))
plt.axvline(x=np.nanmean(x), color='red', linestyle='--', label=f'Mean: {avg_hourly_n_order}') # Mean line
plt.axvline(x=np.nanmedian(x), color='black', linestyle='--', label=f'Median: {mid_hourly_n_order}') # Median line
plt.axvline(x=mode_hourly_n_order, color='black', linestyle='-', label=f'Mode: {mode_hourly_n_order}') # Mode line
plt.legend()
plt.show()
```

### C. Hourly Temperature Distribution

```python
avg_hourly_temp = round(df['temp'].mean(), 2) # Get the mean
mid_hourly_temp = df['temp'].median() # Get the median
mode_hourly_temp = df['temp'].mode()[0] # Get the mode

x = df['temp']
fig = plt.hist(x, bins=10)
plt.title("Hourly Temperature Distribution")
plt.xlabel('Hourly Tempereature')
# plt.xticks(range(0,1100,100))
plt.axvline(x=np.nanmean(x), color='red', linestyle='--', label=f'Mean: {avg_hourly_temp}') # Mean line
plt.axvline(x=np.nanmedian(x), color='black', linestyle='--', label=f'Median: {mid_hourly_temp}') # Median line
plt.axvline(x=mode_hourly_temp, color='black', linestyle='-', label=f'Mode: {mode_hourly_temp}') # Mode line
plt.legend()
plt.show()
```

```python
avg_hourly_wind = round(df['windspeed'].mean(), 2) # Get the mean
mid_hourly_wind = df['windspeed'].median() # Get the median
mode_hourly_wind = df['windspeed'].mode()[0] # Get the mode // Mode "0" artinya apa???

x = df['windspeed']
fig = plt.hist(x, bins=10)
plt.title("Hourly Windspeed Distribution")
plt.xlabel('Hourly Windspeed (km/h)')
plt.axvline(x=np.nanmean(x), color='red', linestyle='--', label=f'Mean: {avg_hourly_wind}') # Mean line
plt.axvline(x=np.nanmedian(x), color='black', linestyle='--', label=f'Median: {mid_hourly_wind}') # Median line
plt.axvline(x=mode_hourly_wind, color='black', linestyle='-', label=f'Mode: {mode_hourly_wind}') # Mode line
plt.legend()
plt.show()
```

### 1.A. Hourly Total Order Distribution and Centrality

```python
# Define the variables
avg_hourly_n_order = round(df['total_order'].mean(), 2) # Get the mean
mid_hourly_n_order = df['total_order'].median() # Get the median
mode_hourly_n_order = df['total_order'].mode()[0] # Get the mode
var_hourly_n_order = round(np.var(df['total_order']),2)
std_hourly_n_order = round(np.std(df['total_order']),2)

# Print the variables
print("Hourly # of Order" )
print("----------------------")
print("Mean:", avg_hourly_n_order)
print("Median:", mid_hourly_n_order)
print("Mode:", mode_hourly_n_order)
print("Var:", var_hourly_n_order)
print("StDev:", std_hourly_n_order)

# Draw the Histogram
x = df['total_order']
fig = plt.hist(x, bins=50)
plt.title("Hourly # of Order(s) Distribution")
plt.xlabel('Hourly # of Order(s)')
plt.xticks(range(0,1100,100))
plt.axvline(x=np.nanmean(x), color='red', linestyle='--', label=f'Mean: {avg_hourly_n_order}') # Mean line
plt.axvline(x=np.nanmedian(x), color='black', linestyle='--', label=f'Median: {mid_hourly_n_order}') # Median line
plt.axvline(x=mode_hourly_n_order, color='black', linestyle='-', label=f'Mode: {mode_hourly_n_order}') # Mode line
plt.legend()
plt.show()
```

## 2. Discrete Variable Analysis

```python
# Create a function to calculate the expected value of discrete variable
def calc_exp_val_disc_var(dataframe, value, weight):
    val = dataframe[value]
    wt = dataframe[weight]
    return round((val * wt).sum() / wt.sum(),2)
```

### 2.A. Hourly # of Order Expected Value

```python
# Group the data by hourly # of order and find the count as the probability weight
grouped_ser = df.groupby('total_order')['total_order'].count()
df_total_order_freq = grouped_ser.reset_index(name='count')

df_total_order_freq
```

```python
# Calculate the expected value of hourly # of order random variable
exp_val_hourly_n_order = calc_exp_val_disc_var(df_total_order_freq, 'total_order', 'count')

print('The expected value of "Hourly # of Order" random variable is:', exp_val_hourly_n_order)
```

### 2.B. Hourly # of Order given the Season

```python
# Group the data by hourly # of order and find the count as the probability weight
grouped_ser = df.groupby(['season', 'total_order'])['total_order'].count()
df_total_order_given_season_freq = grouped_ser.reset_index(name='count')

df_total_order_given_season_freq
```

```python
# Weighted average of value grouped by season
df_total_order_given_season_freq.groupby('season').apply(calc_exp_val_disc_var,'total_order', 'count')
```

### 2.C. Hourly # of Order given the Month

```python
# Group the data by hourly # of order and find the count as the probability weight
grouped_ser = df.groupby(['month', 'total_order'])['total_order'].count()
df_total_order_given_month_freq = grouped_ser.reset_index(name='count')

df_total_order_given_month_freq
```

```python
# Weighted average of value grouped by month
res = df_total_order_given_month_freq.groupby('month').apply(calc_exp_val_disc_var,'total_order', 'count').nlargest(5)
res
```

### 2.D. Expected Value of Hourly # of Order given the Hour
